```python
def seed_torch(seed=42,device=None):
    random.seed(seed)
    os.environ['PYTHONHASHSEED'] = str(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    if device.type == 'cuda':
        torch.cuda.manual_seed(seed)
        torch.cuda.manual_seed_all(seed) # if you are using multi-GPU.
    torch.backends.cudnn.benchmark = False
    torch.backends.cudnn.deterministic = True
    
def list2txt(lines,filepath):
    with open(filepath,"w") as f:
        for line in lines:
            f.write(line+"\n")  
    # 自带文件关闭功能，不需要再写f.close()
```

## cox特征选择

```python
def COX_FeatureSelect(X, y, duration_col, event_col):
    """
    单变量COX回归，然后多变量COX回归。
    """
    from lifelines import CoxPHFitter# 导入模型函数
    cph = CoxPHFitter()# 实例化一个CoxPHFitter对象
    ######### for循环对所有自变量执行单因素Cox回归分析：#########
    unicox_analysis = pd.DataFrame([])# 初始化一个空的表格，接下来用该表格记录所有自变量的单因素Cox回归分析
    for x_name in X.columns:
        df = pd.concat( [
            y,
            X[[x_name]] ],
            axis=1
        )# 子表格
        cph.fit(df, duration_col=duration_col, event_col=event_col)#单变量cox
        unicox_analysis = unicox_analysis.append(cph.summary)# 将cph.summary拼接到unicox_analysis的尾部
    summary_unicox_analysis = unicox_analysis
    # 保留p值小于0.05的那些自变量
    unicox_analysis = unicox_analysis[unicox_analysis['p']<0.05]

    # 多因素cox回归分析需要的数据：
    # 即事件标志列、生存时间列，以及单因素cox回归特征筛选出来的自变量
    Xy_from_unicox = pd.concat( [X.loc[:,unicox_analysis.index], y],axis=1)
    cph = CoxPHFitter()#实例化一个CoxPHFitter对象
    cph.fit(Xy_from_unicox, duration_col='time', event_col='event')#拟合好了多因素Cox回归分析模型
    summary = cph.summary
    features_selected = list(summary[summary['p']<0.05].index)
    summary_multicox_analysis = summary
    return features_selected, summary_unicox_analysis, summary_multicox_analysis

Cox_sel_features, summary_unicox_analysis, summary_multicox_analysis = COX_FeatureSelect( 
    X_train, 
    y_train,
    duration_col='event', event_col='time'
)

summary_unicox_analysis.to_csv(params['filepath_unicox_summary'])
summary_multicox_analysis.to_csv(params['filepath_multicox_summary'])
```

## 嵌入式特征选择（交叉验证寻找最佳k值）

```python
import numpy as np
from sksurv.tree import SurvivalTree
from sklearn.model_selection import GridSearchCV, KFold

def fit_and_score_features(X, y):
    """
    该函数的作用：输入自变量X、结局变量y，该函数将遍历自变量X中的每一个特征，
    建立只有单个特征的生存决策树，然后记录该生存决策树在数据集上的一致性因子。
    """
    n_features = X.shape[1]# 协变量数量等于X表格的列数
    scores = np.empty(n_features)
    
    m = SurvivalTree()# 实例化一个生存决策树
    for j in range(n_features):
        if type(X) is np.ndarray:# 如果输入的X是数组
            Xj = X[:, j:j+1]
        if type(X) is pd.DataFrame:# 如果输入的X是dataframe
            Xj = X.iloc[:, j:j+1]
        m.fit(Xj, y)# 用第j个特征的数据来拟合模型
        scores[j] = m.score(Xj, y)# 计算一致性因子
    return scores


# pipeline（管道）
from sklearn.feature_selection import SelectKBest
from sklearn.pipeline import Pipeline

def GridCV_SelectKBest(model,X_train, y_train_, n_splits=5, **kwargs):
    random_state = kwargs.get('random_state',params['random_state'])
    pipe = Pipeline([
        ('select', SelectKBest(fit_and_score_features)), 
        ('model', model)
    ])# 定义一个（包含有两个步骤）管道


    # 交叉验证网格搜索
    param_grid = kwargs.get('param_grid',{'select__k': np.arange(5, 15)})# 定义超参数的取值范围
    cv = KFold(n_splits=n_splits, random_state=random_state, shuffle=True)# 定义一个交叉验证的迭代器
    gcv = GridSearchCV(pipe, param_grid, return_train_score=True, cv=cv)# 网格搜索，后续超参数寻优时也讲到
    gcv.fit(X_train, y_train_)

    # 赋予最佳超参数并训练模型
    pipe.set_params(**gcv.best_params_)
    pipe.fit(X_train, y_train_)

    # 获得最佳的选择特征以及模型
    transformer, final_estimator = [s[1] for s in  pipe.steps]
    # 最终选择出的特征名称
    sel_features = transformer.get_feature_names_out().tolist()

    return sel_features, final_estimator

```

```python
from sksurv.ensemble import RandomSurvivalForest
RSF = RandomSurvivalForest(n_estimators=100,
                           min_samples_split=3,
                           min_samples_leaf=3,
                           max_features="sqrt",
                           n_jobs=-1,
                           random_state=params['random_state'])# 实例化一个RandomSurvivalForest
# 人为设定的参数——“超参数”
# 学习出来的参数

RSF_sel_features, RSF_final = GridCV_SelectKBest(RSF,X_train, y_train_,n_splits=5,param_grid={'select__k': np.arange(5, 15)},random_state=params['random_state'])
RSF_score_test = RSF_final.score(X_test.loc[:,RSF_sel_features],y_test_)
print(len(RSF_sel_features),RSF_score_test)

# 保存到文件
list2txt(RSF_sel_features,params['filepath_RSF_sel_features'])
```

```python

import sksurv

def get_ML_IBS(est,X_test,y_train,y_test):
    """
    仅适用于sksurv模型求解IBS
    """
    try:
        survs = est.predict_survival_function(X_test,return_array=True)
        times = est.event_times_
        survs = survs[:,(times>=y_test['time'].min())&(times<y_test['time'].max())]
        times = times[(times>=y_test['time'].min())&(times<y_test['time'].max())]
    except:
        surv_funs = est.predict_survival_function(X_test)
        # time_max = min([surv_funs[i].x.max() for i in range(surv_funs.shape[0])])
        times = np.linspace( y_test['time'].min(),y_test['time'].max(),100 )[:-1]
        survs = np.asarray([[fn(t) for t in times] for fn in surv_funs])
    IBS = sksurv.metrics.integrated_brier_score(survival_train=y_train, survival_test=y_test, estimate=survs, times=times)
    return IBS

def get_ML_AUCs(est,X_test,y_train,y_test):
    """
    仅适用于sksurv模型求解AUC
    """
    try:
        cum_hazards = est.predict_cumulative_hazard_function(X_test,return_array=True)
        times = est.event_times_
        cum_hazards = cum_hazards[:,(times>=y_test['time'].min())&(times<y_test['time'].max())]
        times = times[(times>=y_test['time'].min())&(times<y_test['time'].max())]
    except:
        cum_hazard_funs = est.predict_cumulative_hazard_function(X_test)
        times = np.linspace( y_test['time'].min(),y_test['time'].max(),100 )[:-1]
        cum_hazards = np.asarray([[fn(t) for t in times] for fn in cum_hazard_funs])
    aucs,mean_auc = sksurv.metrics.cumulative_dynamic_auc(
        survival_train = y_train, 
        survival_test = y_test,
        estimate = cum_hazards, 
        times = times, 
        tied_tol=1e-08)
    aucs = aucs[~np.isnan(aucs)]
    mean_auc = aucs.mean()
    return aucs,mean_auc

# """
# X_test_ = X_test.loc[:,feature_sels['CoxBoost']]
# IBS_RSF = get_ML_IBS(CoxBoost,X_test_.values,y_test_)
# aucs,mean_auc = get_ML_AUCs(CoxBoost,X_test_.values,y_train_,y_test_)
# print(IBS_RSF,aucs,mean_auc)
# """
```

```python
from sksurv.util import Surv
def y_pycox_to_y_sksurv(y_pycox):
    y_sksurv = pd.DataFrame(columns=['time','event'])
    y_sksurv['time'] = y_pycox[0].T
    y_sksurv['event'] = y_pycox[1].T
    y_sksurv = Surv.from_dataframe( event='event', time='time', data=y_sksurv)
    return y_sksurv

def get_Deep_AUCs(DeepSurv,X_test_deep,y_train_deep,y_test_deep):
    """
    仅适用于pycox库DeepSurv模型求解AUC
    """
    _ = DeepSurv.compute_baseline_hazards()# 先计算baseline harzard（基准风险函数）
    cum_hazards = DeepSurv.predict_cumulative_hazards(X_test_deep)# 计算生存函数
    cum_hazards = cum_hazards[ (cum_hazards.index>=y_test_deep[0].min())&(cum_hazards.index<y_test_deep[0].max()) ]
    cum_hazards = cum_hazards.T
    times = cum_hazards.columns.values
    cum_hazards = cum_hazards.values
    # 将因变量从pycox格式转为sksurv格式
    y_train_sksurv = y_pycox_to_y_sksurv(y_train_deep)
    y_test_sksurv = y_pycox_to_y_sksurv(y_test_deep)
    AUCs,meanAUC = sksurv.metrics.cumulative_dynamic_auc(
        survival_train = y_train_sksurv, 
        survival_test = y_test_sksurv,
        estimate = cum_hazards, 
        times = times, 
        tied_tol=1e-08)
    notna = ~np.isnan(AUCs)
    AUCs = AUCs[notna]
    times = times[notna]
    meanAUC = AUCs.mean()
    return (AUCs,times),meanAUC

def test_DeepSurv(DeepSurv, X_test_deep, y_test_deep, y_train_deep):
    # 预测测试集上的生存函数
    _ = DeepSurv.compute_baseline_hazards()# 先计算baseline harzard（基准风险函数）
    surv = DeepSurv.predict_surv_df(X_test_deep)# 计算生存函数

    # 构建EvalSurv对象
    durations_test = y_test_deep[0]
    events_test = y_test_deep[1]
    ev = EvalSurv(surv, durations=durations_test, events=events_test, censor_surv='km')
    # c-index
    CIndex = ev.concordance_td()
    # IBS
    times = np.linspace(durations_test.min(), durations_test.max(), 100)
    IBS = ev.integrated_brier_score(times)
    # AUCs
    AUCs,meanAUC = get_Deep_AUCs(DeepSurv,X_test_deep,y_train_deep,y_test_deep)
    return CIndex,IBS,AUCs,meanAUC

# """
# CIndex,IBS,AUCs,meanAUC = test_DeepSurv(DeepSurv, X_train_deep, y_train_deep, y_train_deep)
# print(CIndex,IBS,meanAUC)
# CIndex,IBS,AUCs,meanAUC = test_DeepSurv(DeepSurv, X_test_deep, y_test_deep, y_train_deep)
# print(CIndex,IBS,meanAUC)
# """
```

```python
CIndex_test = pd.DataFrame(index=list(feature_sels.keys()),columns=algs)
IBS_test = pd.DataFrame(index=list(feature_sels.keys()),columns=algs)
AUC_test = pd.DataFrame(index=list(feature_sels.keys()),columns=algs)
CIndex_train = pd.DataFrame(index=list(feature_sels.keys()),columns=algs)
IBS_train = pd.DataFrame(index=list(feature_sels.keys()),columns=algs)
AUC_train = pd.DataFrame(index=list(feature_sels.keys()),columns=algs)
for feature_sel_method in feature_sels:
    sel_features = feature_sels[feature_sel_method]
    X_train_ = X_train.loc[:,sel_features]
    X_test_ = X_test.loc[:,sel_features]
    for alg in algs:
        print(feature_sel_method, alg, ':')
        if alg=='Cox':
            """
            model = CoxPHFitter()
            duration_col, event_col = 'event','time'
            train_df = pd.concat([X_train_, y_train],axis=1)
            test_df = pd.concat([X_test_, y_test],axis=1)
            model.fit(train_df, duration_col=duration_col, event_col=event_col)
            CIndex_test.loc[feature_sel_method,alg] = model.score(test_df,scoring_method='concordance_index')
            """
            model = sksurv.linear_model.CoxPHSurvivalAnalysis(alpha=0, ties='breslow', n_iter=100, tol=1e-09, verbose=0)
            model.fit(X_train_,y_train_)
            # c-index
            CIndex_test.loc[feature_sel_method,alg] = model.score(X=X_test_,y=y_test_)
            CIndex_train.loc[feature_sel_method,alg] = model.score(X=X_train_,y=y_train_)
            # IBS, mean_AUC
            IBS_test.loc[feature_sel_method,alg] = get_ML_IBS(model,X_test_,y_train_,y_test_)
            _,AUC_test.loc[feature_sel_method,alg] = get_ML_AUCs(model,X_test_,y_train_,y_test_)
            IBS_train.loc[feature_sel_method,alg] = get_ML_IBS(model,X_train_,y_train_,y_train_)
            _,AUC_train.loc[feature_sel_method,alg] = get_ML_AUCs(model,X_train_,y_train_,y_train_)

        
        elif alg=='RSF':
            model = RandomSurvivalForest(
                n_estimators=100,
                min_samples_split=10,
                min_samples_leaf=15,
                max_features="sqrt",
                n_jobs=-1,
                random_state=params['random_state'])
            model.fit(X=X_train_,y=y_train_)
            # c-index
            CIndex_test.loc[feature_sel_method,alg] = model.score(X=X_test_,y=y_test_)
            CIndex_train.loc[feature_sel_method,alg] = model.score(X=X_train_,y=y_train_)
            # IBS, mean_AUC
            IBS_test.loc[feature_sel_method,alg] = get_ML_IBS(model,X_test_,y_train_,y_test_)
            _,AUC_test.loc[feature_sel_method,alg] = get_ML_AUCs(model,X_test_,y_train_,y_test_)
            IBS_train.loc[feature_sel_method,alg] = get_ML_IBS(model,X_train_,y_train_,y_train_)
            _,AUC_train.loc[feature_sel_method,alg] = get_ML_AUCs(model,X_train_,y_train_,y_train_)


        elif alg=='CoxBoost':
            model = GradientBoostingSurvivalAnalysis(loss="coxph",learning_rate=0.01)
            model.fit(X=X_train_,y=y_train_)
            # c-index
            CIndex_test.loc[feature_sel_method,alg] = model.score(X=X_test_,y=y_test_)
            CIndex_train.loc[feature_sel_method,alg] = model.score(X=X_train_,y=y_train_)
            # IBS, meanAUC
            IBS_test.loc[feature_sel_method,alg] = get_ML_IBS(model,X_test_,y_train_,y_test_)
            _,AUC_test.loc[feature_sel_method,alg] = get_ML_AUCs(model,X_test_,y_train_,y_test_)
            IBS_train.loc[feature_sel_method,alg] = get_ML_IBS(model,X_train_,y_train_,y_train_)
            _,AUC_train.loc[feature_sel_method,alg] = get_ML_AUCs(model,X_train_,y_train_,y_train_)
        
        elif alg=='DeepSurv':
            # 数据格式整理
            # 将结局变量整理成pycox要求的格式: 生存时间数组和事件标志数组组成的元组
            y_train_deep = ( y_train['time'].values, y_train['event'].values )
            y_test_deep = ( y_test['time'].values, y_test['event'].values )
            # 将自变量整理成pycox要求的格式
            X_train_deep = X_train_.values.astype('float32')# 32位浮点数的数组
            X_test_deep = X_test_.values.astype('float32')# 32位浮点数的数组
            # 训练
            model = train_DeepSurv(X_train_deep,y_train_deep)
            # 测试
            # c-index, IBS, meanAUC
            CIndex,IBS,AUCs,meanAUC = test_DeepSurv(model, X_test_deep, y_test_deep, y_train_deep)
            CIndex_test.loc[feature_sel_method,alg] = CIndex
            IBS_test.loc[feature_sel_method,alg] = IBS
            AUC_test.loc[feature_sel_method,alg] = meanAUC
            CIndex,IBS,AUCs,meanAUC = test_DeepSurv(model, X_train_deep, y_train_deep, y_train_deep)
            CIndex_train.loc[feature_sel_method,alg] = CIndex
            IBS_train.loc[feature_sel_method,alg] = IBS
            AUC_train.loc[feature_sel_method,alg] = meanAUC
        # 保存训练好的模型对象
        trained_models['{}_{}'.format(feature_sel_method,alg)] = model
        print('Test-set CIndex: {}'.format(CIndex_test.loc[feature_sel_method,alg]))
```

```python
def get_DCA_coords(y_true,y_proba):
    """
    Return coordinates of DCA curve points
    输入：
        y_proba: 预测的阳性概率值。
        y_true: 真实值，0表示阴性，1表示阳性。
    输出：
        pt_arr: 阈值，即DCA曲线横坐标，数组格式。
        net_bnf_arr: 净收益，即DCA曲线纵坐标，数组格式。
    """
    pt_arr = [] # 定义一个列表，用于记录阈值
    net_bnf_arr = [] # 定义一个列表，用于记录净收益
    y_proba = y_proba.ravel()# 展平
    n = len(y_proba)# 样本数量
    for i in range(0,100,1):
        pt = i/100 # 概率阈值
        #compute TP FP
        y_pred = (y_proba>pt).astype(int) # 大于阈值预测为阳性
        TP = np.sum( y_true*np.round(y_pred) )
        FP = np.sum( (1 - y_true) * np.round(y_pred) )
        net_bnf = ( TP-FP* pt/(1-pt) )/n
        pt_arr.append(pt)
        net_bnf_arr.append(net_bnf)
    
    return np.array(pt_arr), np.array(net_bnf_arr)
```
