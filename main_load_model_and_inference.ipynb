# define EnsModel

```python
class EnsModel(object):
    def __init__(self,ens_model,name_topk_models,models,features):
        """
        ens_model: ensembled model.
        name_topk_models: list, names of model to be ensembled.
        models: dict: trained models to be ensembled.
        features: dict: list of list, features for each model.
        """
        self.ens_model = ens_model
        self.name_topk_models = name_topk_models
        self.models = models
        self.features = features
        
        # set attributes
        feat_names = []
        for v in features.values():
            feat_names += v
        self.feature_names_in_ = np.array(list(set(feat_names)))
        self.n_features_in_ = sum([len(v) for v in self.features.values()])
        self.baseline_survival_ = self.ens_model.baseline_survival_
        self.cum_baseline_hazard_ = self.ens_model.cum_baseline_hazard_
        self.coef_ = self.ens_model.coef_
        
        
    def predict_submodels_log_hazard_ratio(self,X_df):
        log_hazard_ratio = dict([])
        for name in self.name_topk_models:
            feature_sel_method, alg = name.split('_')
            sel_features = self.features[feature_sel_method]
            log_hazard_ratio[name] = self.models[name].predict(X_df.loc[:,sel_features])
        log_hazard_ratio = pd.DataFrame(log_hazard_ratio)
        return log_hazard_ratio
    
    def predict(self,X_df):
        """
        predict(X_df) method of sksurv.linear_model.coxph.CoxPHSurvivalAnalysis instance
        Predict risk scores.

        Parameters
        ----------
        X_df : dataframe, shape = (n_samples, n_features)
            Data matrix.

        Returns
        -------
        risk_score : array, shape = (n_samples,)
            Predicted risk scores.
        """
        submodels_log_hazard_ratio = self.predict_submodels_log_hazard_ratio(X_df)
        risk_scores = self.ens_model.predict( submodels_log_hazard_ratio )
        return risk_scores
    
    def predict_survival_function(self,X_df):
        """
        predict_survival_function(X_df) method of sksurv.linear_model.coxph.CoxPHSurvivalAnalysis instance
        Predict survival function.

        Parameters
        ----------
        X_df : dataframe, shape = (n_samples, n_features)
            Data matrix.

        Returns
        -------
        survival : ndarray of :class:`sksurv.functions.StepFunction`, shape = (n_samples,)
            Predicted survival functions.
        """
        submodels_log_hazard_ratio = self.predict_submodels_log_hazard_ratio(X_df)
        surv_funs = self.ens_model.predict_survival_function( submodels_log_hazard_ratio )
        return surv_funs
    
    def predict_cumulative_hazard_function(self,X_df):
        """
        predict_cumulative_hazard_function(X) method of sksurv.linear_model.coxph.CoxPHSurvivalAnalysis instance.
        Input:
            X_df : dataframe, shape = (n_samples, n_features)
        Return:
            cum_hazard : ndarray of :class:`sksurv.functions.StepFunction`, shape = (n_samples,) Predicted cumulative hazard functions.
        """
        submodels_log_hazard_ratio = self.predict_submodels_log_hazard_ratio(X_df)
        surv_funs = self.ens_model.predict_cumulative_hazard_function( submodels_log_hazard_ratio )
        return surv_funs
    
    def score(self,X_df,y):
        """
        score(X, y) method of sksurv.linear_model.coxph.CoxPHSurvivalAnalysis instance
        Returns the concordance index of the prediction.

        Parameters
        ----------
        X_df : dataframe, shape = (n_samples, n_features)
            Test samples.

        y : structured array, shape = (n_samples,)
            A structured array containing the binary event indicator
            as first field, and time of event or time of censoring as
            second field.

        Returns
        -------
        cindex : float
            Estimated concordance index.
        """
        submodels_log_hazard_ratio = self.predict_submodels_log_hazard_ratio(X_df)
        scores = self.ens_model.score( submodels_log_hazard_ratio, y )
        return scores
```
